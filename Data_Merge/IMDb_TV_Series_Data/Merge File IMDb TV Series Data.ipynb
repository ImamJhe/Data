```python
import pandas as pd
import os
```

```python
# Read a specific CSV file as df
df = pd.read_csv("D:\Imam\Belajar\Python\Data_Science\Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\Sales_April_2019.csv")

# Get a list of all files in a directory
files = [file for file in os.listdir('D:\Imam\Belajar\Python\Data_Science\Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data')]

# Create an empty DataFrame to store all data
all_months_data = pd.DataFrame()

# Iterate over the files and concatenate the data
for file in files:
    # Read each CSV file and append the data to the DataFrame
    df = pd.read_csv("D:\Imam\Belajar\Python\Data_Science\Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data/"+file, sep=';')
    all_months_data = pd.concat([all_months_data, df])
    
# Save the combined data to a new CSV file
all_months_data.to_csv("All_data.csv", index=False)
```
