```python
import pandas as pd
import os
import re
```

```python
# Get a list of all files in a directory
files = [file for file in os.listdir('D:\Imam\Belajar\Python\Data_Merge\Data')]

# Create an empty DataFrame to store all data
all_data = pd.DataFrame()

# Iterate over the files and concatenate the data
for file in files:
    # Read each CSV file and append the data to the DataFrame
    df = pd.read_csv('D:\Imam\Belajar\Python\Data_Merge\Data/'+file, sep=',')
    all_data = pd.concat([all_data, df])
    
# Save the combined data to a new CSV file
all_data.to_csv("All_data.csv", index=False)
```

```python
# Read Data
df_all_data = pd.read_csv("All_data.csv")
```

```python
# Print Data
df = df_all_data.copy()
```

```python
df.isna().sum()
```

```python
df['Certificate'] = df['Certificate'].fillna('-')
df['Gross Revenue'] = df['Gross Revenue'].fillna('-')
df['Runtime'] = df['Runtime'].fillna('-')
df.dropna(subset = ['Release Year'], inplace = True)
df.dropna(subset = ['Cast'], inplace = True)
```

```python
df.isna().sum()
```

```python
def year(item):
    match = re.search(r'\b\d{4}\b', item)
    if match:
        return match.group()
    else:
        return None
```

```python
df['Release Year'] = list(map(year, df['Release Year']))
```
